```python
bobot = {
    "hadir": 10,
    "tugas": 20,
    "uts": 30,
    "uas": 40
}
def hitung_nilai(sks, hadir, tugas, uts, uas, bobot = bobot):
    max_hadir = 0
    if sks == 2:
        max_hadir = 14
    if sks == 3:
        max_hadir = 21
    if not max_hadir:
        return "SKS tidak valid"

    # Kehadiran
    hadir = hadir/max_hadir * bobot["hadir"]
    # Tugas, uts, uas
    tugas = tugas/100 * bobot["tugas"]
    uts = uts/100 * bobot["uts"]
    uas = uas/100 * bobot["uas"]
    
    return round(hadir + tugas + uts + uas, 2)
```

```python
# Fungsi untuk menghitung grade berdasarkan nilai akhir
def hitung_grade(nilai):
    if nilai >= 85:
        return "A"
    elif nilai >= 70:
        return "B"
    elif nilai >= 60:
        return "C"
    elif nilai >= 50:
        return "D"
    else:
        return "E"

# Menggunakan fungsi hitung_nilai untuk menghitung nilai dan grade setiap mata kuliah
bobot = {
    "hadir": 10,
    "tugas": 20,
    "uts": 30,
    "uas": 40
}

def hitung_nilai(sks, hadir, tugas, uts, uas, bobot=bobot):
    max_hadir = 0
    if sks == 2:
        max_hadir = 14
    if sks == 3:
        max_hadir = 21
    if not max_hadir:
        return "SKS tidak valid"
    
    # Kehadiran
    hadir = hadir / max_hadir * bobot["hadir"]
    # Tugas, UTS, UAS
    tugas = tugas / 100 * bobot["tugas"]
    uts = uts / 100 * bobot["uts"]
    uas = uas / 100 * bobot["uas"]
    
    # Total nilai
    total_nilai = round(hadir + tugas + uts + uas, 2)
    
    return total_nilai, hitung_grade(total_nilai)

# Menghasilkan hasil akhir untuk setiap siswa
results = []
for student in students:
    rangkuman_nilai = []
    for mk in student["hasil"]:
        nilai_akhir, grade = hitung_nilai(
            mk["sks"], mk["hadir"], mk["nilai"]["tugas"], mk["nilai"]["uts"], mk["nilai"]["uas"]
        )
        rangkuman_nilai.append({
            "mata_kuliah": mk["mata_kuliah"],
            "nilai": f"{nilai_akhir} : 0 - 100",
            # "Grade": f"{grade}: A | B | C | D | E"
        })
    
    results.append({
        "nim": student["nim"],
        "nama": student["nama"],
        "rangkuman_nilai": rangkuman_nilai
    })

# Menampilkan hasil akhir
for result in results:
    print("NIM:", result["nim"])
    print("Nama:", result["nama"])
    for nilai in result["rangkuman_nilai"]:
        print(f"Mata Kuliah: {nilai['mata_kuliah']}, Nilai: {nilai['nilai']},")
    print("----------------------------")
```
