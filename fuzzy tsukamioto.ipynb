```python
import numpy as np

def turun(x, a, b):
    if x <= a:
        return 1
    elif a < x <= b:
        return (b - x) / (b - a)
    else:
        return 0

def tetap(x, a, b, c):
    if a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    else:
        return 0

def naik(x, a, b):
    if x <= a:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    else:
        return 1

def sedikit(x, a, b):
    return turun(x, a, b)

def sedang(x, a, b, c):
    return tetap(x, a, b, c)

def banyak(x, a, b):
    return naik(x, a, b)

def berkurang(y, a, b):
    return turun(y, a, b)

def bertambah(y, a, b):
    return naik(y, a, b)

# Nilai input (permintaan dan persediaan)
permintaan = 4000  # Contoh input
persediaan = 600   # Contoh input

# Membership function untuk permintaan (turun, tetap, naik)
μ_turun = turun(permintaan, 0, 3000)
μ_tetap = tetap(permintaan, 2000, 3000, 5000)
μ_naik = naik(permintaan, 4000, 7000)

# Membership function untuk persediaan (sedikit, sedang, banyak)
μ_sedikit = sedikit(persediaan, 0, 400)
μ_sedang = sedang(persediaan, 200, 500, 800)
μ_banyak = banyak(persediaan, 600, 1000)

# Rules (Inferensi)
rules = [
    (min(μ_turun, μ_sedikit), 'bertambah'),
    (min(μ_turun, μ_sedang), 'berkurang'),
    (min(μ_turun, μ_banyak), 'berkurang'),
    (min(μ_tetap, μ_sedikit), 'bertambah'),
    (min(μ_tetap, μ_sedang), 'berkurang'),
    (min(μ_tetap, μ_banyak), 'berkurang'),
    (min(μ_naik, μ_sedikit), 'bertambah'),
    (min(μ_naik, μ_sedang), 'bertambah'),
    (min(μ_naik, μ_banyak), 'berkurang'),
]

# Produksi (defuzzifikasi Tsukamoto)
def tsukamoto_defuzz(rules):
    numerator = 0
    denominator = 0
    
    for strength, output in rules:
        if output == 'bertambah':
            y = 9000  # Nilai produksi maksimum
        elif output == 'berkurang':
            y = 1000  # Nilai produksi minimum
        
        numerator += strength * y
        denominator += strength
    
    return numerator / denominator if denominator != 0 else 0

# Hitung hasil defuzzifikasi
produksi = tsukamoto_defuzz(rules)

# Tampilkan hasil
print(f"Permintaan: {permintaan}")
print(f"Persediaan: {persediaan}")
print(f"Produksi yang disarankan: {produksi:.2f}")
```
