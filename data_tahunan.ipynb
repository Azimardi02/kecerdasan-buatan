```python
# Data rekap yang diberikan
data_rekap = [
    {"tahun": 1995, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 1997, "penjualan": 320, "operasional": 120, "pajak": 70, "pendapatan_lain": 105},
    {"tahun": 1991, "penjualan": 200, "operasional": 75, "pajak": 35, "pendapatan_lain": 85},
    {"tahun": 1992, "penjualan": 320, "operasional": 170, "pajak": 80, "pendapatan_lain": 235},
    {"tahun": 2001, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 2003, "penjualan": 500, "operasional": 350, "pajak": 100, "pendapatan_lain": 55},
    {"tahun": 2005, "penjualan": 750, "operasional": 450, "pajak": 200, "pendapatan_lain": 155}
]

# Mengambil data tahun dan penjualan, lalu mengurutkan berdasarkan tahun
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}
tahun = list(data_map.keys())
tahun.sort()

# Mendapatkan data penjualan berdasarkan urutan tahun
penjualan = [data_map[_tahun] for _tahun in tahun]

# Menghitung laba untuk setiap tahun
laba = [
    data["penjualan"] + data["pendapatan_lain"] - data["pajak"] - data["operasional"]
    for data in sorted(data_rekap, key=lambda x: x["tahun"])
]

# Mengumpulkan data lain seperti pajak, operasional, dan pendapatan_lain berdasarkan urutan tahun
pajak = [data["pajak"] for data in sorted(data_rekap, key=lambda x: x["tahun"])]
operasional = [data["operasional"] for data in sorted(data_rekap, key=lambda x: x["tahun"])]
pendapatan_lain = [data["pendapatan_lain"] for data in sorted(data_rekap, key=lambda x: x["tahun"])]

# Menyusun data dalam format yang diharapkan
data_graph = {
    "tahun": tahun,
    "penjualan": penjualan,
    "pajak": pajak,
    "operasional": operasional,
    "pendapatan_lain": pendapatan_lain,
    "laba": laba
}

# Menampilkan hasil data_graph
print(data_graph)

# Plotting data
from matplotlib import pyplot as plt

plt.plot(data_graph["tahun"], data_graph["penjualan"], "o-", label="penjualan")
plt.plot(data_graph["tahun"], data_graph["pajak"], "o-", label="pajak")
plt.plot(data_graph["tahun"], data_graph["laba"], "o-", label="laba")

plt.title("Data Rekapitulasi Tahunan")
plt.legend(loc="upper left")
plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan Rp.")
plt.show()
```
