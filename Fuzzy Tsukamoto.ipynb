```python
import numpy as np

def permintaan_turun(x):
    if x <= 1000:
        return 1
    elif 1000 < x <= 3000:
        return (3000 - x) / 2000
    else:
        return 0

def permintaan_tetap(x):
    if 2000 <= x <= 4000:
        return (x - 2000) / 2000
    elif 4000 < x <= 5000:
        return (5000 - x) / 1000
    else:
        return 0

def permintaan_naik(x):
    if x <= 4000:
        return 0
    elif 4000 < x <= 6000:
        return (x - 4000) / 2000
    else:
        return 1

def persediaan_sedikit(x):
    if x <= 200:
        return 1
    elif 200 < x <= 400:
        return (400 - x) / 200
    else:
        return 0

def persediaan_sedang(x):
    if 200 <= x <= 500:
        return (x - 200) / 300
    elif 500 < x <= 800:
        return (800 - x) / 300
    else:
        return 0

def persediaan_banyak(x):
    if x <= 600:
        return 0
    elif 600 < x <= 1000:
        return (x - 600) / 400
    else:
        return 1

def produksi_berkurang(mu):
    return 5000 - mu * (5000 - 1000)

def produksi_bertambah(mu):
    return 1000 + mu * (9000 - 1000)

# Rule Fuzzy Tsukamoto
def fuzzy_tsukamoto(permintaan, persediaan):
    rules = []

    # Aturan sesuai rule yang diberikan
    rules.append(min(permintaan_turun(permintaan), persediaan_sedikit(persediaan)))
    rules.append(min(permintaan_turun(permintaan), persediaan_sedang(persediaan)))
    rules.append(min(permintaan_turun(permintaan), persediaan_banyak(persediaan)))
    rules.append(min(permintaan_tetap(permintaan), persediaan_sedikit(persediaan)))
    rules.append(min(permintaan_tetap(permintaan), persediaan_sedang(persediaan)))
    rules.append(min(permintaan_tetap(permintaan), persediaan_banyak(persediaan)))
    rules.append(min(permintaan_naik(permintaan), persediaan_sedikit(persediaan)))
    rules.append(min(permintaan_naik(permintaan), persediaan_sedang(persediaan)))
    rules.append(min(permintaan_naik(permintaan), persediaan_banyak(persediaan)))

    # Produksi berdasarkan rule
    produksi = [
        produksi_bertambah(rules[0]),
        produksi_berkurang(rules[1]),
        produksi_berkurang(rules[2]),
        produksi_bertambah(rules[3]),
        produksi_berkurang(rules[4]),
        produksi_berkurang(rules[5]),
        produksi_bertambah(rules[6]),
        produksi_bertambah(rules[7]),
        produksi_berkurang(rules[8]),
    ]

    # Perhitungan weighted average
    numerator = sum(rules[i] * produksi[i] for i in range(len(rules)))
    denominator = sum(rules)

    if denominator == 0:
        return 0
    return numerator / denominator

# Input Nilai Permintaan dan Persediaan
permintaan = float(input("Masukkan nilai permintaan: "))
persediaan = float(input("Masukkan nilai persediaan: "))

# Hitung output produksi
output_produksi = fuzzy_tsukamoto(permintaan, persediaan)
print(f"Produksi yang dihasilkan adalah: {output_produksi:.2f}")
```
